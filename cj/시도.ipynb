# 인구소멸지역 분석

## 지방소멸위험지수: ‘20-39세 여성인구 수' / '65세 이상 고령인구 수’

 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('./시도.csv',thousands=',')
df.head()
```

```python
df.info()
```

```python
# 전국 행제거
df = df[df['  시도  '] != '전국']
```

```python
plt.figure(figsize=(10, 6))
x = df['  시도  ']

# 20-39세 여성인구 그래프 그리기
plt.bar(x, df['20-39세 여성인구'], label='20-39세 여성인구', color='blue')

# 65세 이상 인구 그래프 그리기
plt.bar(x, df['65세 이상 인구'], label='65세 이상 인구', bottom=df['20-39세 여성인구'], color='orange')

plt.title('20-39세 여성인구 및 65세 이상 인구')
plt.xlabel('시도')
plt.ylabel('인구수(단위:백만)')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df = df.sort_values(by='지방소멸위험지수', ascending=True)
plt.rc('font', family='AppleGothic') #맥
plt.bar(df['  시도  '], df['지방소멸위험지수'] ,color='lightblue')
plt.xlabel('시도')
plt.ylabel('지방소멸위험지수')
plt.title('시도별 인구소멸 위험 지수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
